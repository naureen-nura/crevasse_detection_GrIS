```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
Time = [2015,2016,2017,2018,2019,2021,2022,2023]
Crevasse = [65.1776,69.087,64.2659,69.9895,66.8596,73.4819,71.3262,70.7848]
#Portion = [0.103597005208333,10.9810829162598,10.2147897084554,11.1245314280192,
#           10.6270472208659,11.6796334584554,11.3369941711426,11.2509409586589]
ax = sns.regplot(x=Time, y=Crevasse, scatter=False, label= "Line of Fit", color='green')
plt.errorbar(x=Time, y=Crevasse, yerr=1.12, label= "Crevassed Area", ecolor='orange')
plt.legend(loc='lower right')
#ax.scatter(2020, 124.0498)
ax.set(xlabel="Years (January Month)",ylabel="Total Crevassed Area ($km^2$)")
ax.xaxis.get_label().set_fontsize(14)
ax.yaxis.get_label().set_fontsize(14)
```
